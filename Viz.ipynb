```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
radar1 = pd.read_csv("./datasets/radar/2017-13.csv")
```

```python
ave_0 = np.sum(np.arange(34) * radar1.values[:,2:], axis=1) /2016
```

```python
radar1['ave_0'] = ave_0
```

```python
ave_1 = np.sum(np.arange(1,34) * radar1.values[:,3:-1], axis=1) /np.sum(radar1.values[:,3:-1], axis=1)
```

```python
radar1['ave_1'] = ave_1
```

```python
sns.set()
```

```python
viz = radar1.pivot("lat", 'long', 'ave_0')
f, ax = plt.subplots(figsize=(10, 10))
sns.heatmap(viz)
```

```python
viz = radar1.pivot("lat", 'long', 'ave_1')
f, ax = plt.subplots(figsize=(10, 10))
sns.heatmap(viz)
```
